
Programming Questions on Operators in Python


```python
# Q1. Arithmetic Operators
# Write a Python program that takes two numbers from the user and performs all arithmetic operations ( +, -, , /, %, //, * ). Print the results.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Sum : ",a+b)
print("minus : ",a-b)
print("multiply : ",a*b)
print("division : ",a/b)
print("floor division : ",a//b)
print("modulus : ",a%b)
print("power : ",a**b)
```

```python
# Q2. Assignment Operators
# Write a Python program that initializes a variable with a value and performs the following assignment operations: +=, -=, *=, /=, %=. Print the value of the variable after each operation.
a=int(input("Enter first number: "))
a += 1
print("After += :", a)

a -= 1
print("After -= :", a)

a *= 2
print("After *= :", a)

a /= 2
print("After /= :", a)

a %= 3
print("After %= :", a)
```

```python
# Q3. Comparison Operators
# Write a Python program to compare two numbers input by the user. Use ==, !=, >, <, >=, and <= operators and print the result of each comparison.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("a == b :", a == b)
print("a != b :", a != b)
print("a > b  :", a > b)
print("a < b  :", a < b)
print("a >= b :", a >= b)
print("a <= b :", a <= b)
```

```python
# Q4. Logical Operators
# Take three numbers from the user. Write a Python program to check if the first number is greater than both the second and third numbers using logical operators (and, or, not).
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
if(a>b and a>c):
    print(f"{a} is largest number")
else:
    print(f"{a} is not a largest number")
```

```python
# Q5. Bitwise Operators
# Write a Python program to take two integers and show the result of all bitwise operators: &, |, ^, ~, <<, and >>.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("a & b  =", a & b)    # AND
print("a | b  =", a | b)    # OR
print("a ^ b  =", a ^ b)    # XOR
print("~a     =", ~a)       # NOT
print("~b     =", ~b)       # NOT
print("a << 1 =", a << 1)   # Left Shift
print("a >> 1 =", a >> 1)   # Right Shift
```

```python
# Q6. Identity Operators
# Create two lists with the same values. Use the is and is not operators to check if they refer to the same object in memory. Print the result.
list1=[1,3,4]
list2=[1,3,4]
print(list1 is list2)
print(list1 is not list2)
```

```python
# Q7. Membership Operators
# Create a list of numbers. Take an input from the user and check if the number is in the list or not using the in and not in operators.
list1=[6,3,8,1,67,"k",23,87]
a=int(input("Enter a number"))
print(a in list1)
print(a not in list1)
```

```python
# Q8. Calculator Using Operators
# Create a simple calculator that takes two numbers and an operator (+, -, , /, *, //, %) as input from the user and performs the selected operation.
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, //, %): ")

if op == "+":
    print("Result =", a + b)

elif op == "-":
    print("Result =", a - b)

elif op == "*":
    print("Result =", a * b)

elif op == "/":
    print("Result =", a / b)

elif op == "//":
    print("Result =", a // b)

elif op == "%":
    print("Result =", a % b)

else:
    print("Invalid operator")
```
